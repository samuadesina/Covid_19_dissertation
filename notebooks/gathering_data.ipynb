```python
# check all the rows with the Australia location key
data[data["location_key"] == "AU"]
```

```python
# create a new column named column and map all the rows to corresponding country,
# ... based on the location key
data["country"] = data["location_key"].map(country_mapping)
```

```python
# visuaize the transformmed data
data
```

```python
# reordering the columns of the data to take country to the second index

# converr the columns to list
columns = data.columns.to_list()
# reorder the converted columns based on indexing
columns.insert(1, columns.pop(-1))
# map the new columns to the data, to create the reorder
data = data[columns]
```

```python
# take note of countries to be studied
countries = ["United States of America", "Nigeria", "South Africa", "United Kingdom of Great Britain and Northern Ireland",
             "Canada", "Australia", "New Zealand", "Germany", "France", "China",
             "Saudi Arabia", "Brazil", "Argentina"]

# check for the length of the countries above
print(len(countries))

# collect only rows that have thier country names as the countries above
poc = data[data["country"].isin(countries)]
# process the date column to datetime object for easier manipulation
poc["date"] = pd.to_datetime(poc["date"])

# check the length of all the countries in the transformed dataframe
print(len(poc["country"].unique()))
```

```python
# read vaccination data from google open health repository
vacc = pd.read_csv("https://storage.googleapis.com/covid19-open-data/v3/latest/vaccinations.csv")
# process the date column to datetime object for easier manipulation
vacc["date"] = pd.to_datetime(vacc["date"])
# sort the dataframe by date and display
vacc.sort_values(by="date")
```

```python
# read hospitality data from google open health repository
hosp = pd.read_csv("https://storage.googleapis.com/covid19-open-data/v3/hospitalizations.csv")

# Challenge #1: Inconsistent (or false) year values in the date column
# Year values was replaced with a supposed value for the inconsistent ones
hosp["date"] = hosp["date"].str.replace("0022", "2022").str.replace("0202", "2020").str.replace("1202", "2021").str.replace("0221", "2021")
# process the date column to datetime object for easier manipulation
hosp['date'] = pd.to_datetime(hosp['date'], format='%Y-%m-%d')
# sort the dataframe by date
sort_hosp = hosp.sort_values(by="date")

# collect from January, 2020 to December, 2022
# Filter the dataframe based on the date range
start_date = pd.to_datetime('2020-01-01')
end_date = pd.to_datetime('2022-12-31')

# filter the dataframe based on the values above
filtered_sort_hosp = sort_hosp[(sort_hosp['date'] >= start_date) & (sort_hosp['date'] <= end_date)]

# display filtered dataframe
filtered_sort_hosp
```

```python
# merging the transformed epidomological data with vaccination data on the location_key columns
merged_df = (
    poc.merge(
        vacc[["date", "location_key", "new_persons_vaccinated",
              "new_persons_vaccinated_pfizer",
              "new_persons_fully_vaccinated_janssen",
              "new_persons_vaccinated_sinovac"]],
        on="location_key")
)
# display merge results
merged_df
```

```python
# rename columns based on generated names throughout the merge
merged_df.rename(columns={'date_x': 'date'}).merge(filtered_sort_hosp[["date", "location_key", "new_hospitalized_patients"]],
            on="date")

# ... and drop irrelevant columns
# .drop(['location_key_y'], axis=1)
# .rename(columns={'location_key_x': 'location_key'})
```
