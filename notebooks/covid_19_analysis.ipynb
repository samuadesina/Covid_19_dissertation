`Matplotlib` is a plotting library that provides low-level control over the appearance of plots. It is the foundation on which `seaborn` is built and can be used to create a wide variety of plots, including line plots, scatter plots, bar plots, and more.

## Dataset Information & Statistics

```python
data.info()
```

This is a DataFrame with 12 columns and 12842 rows. Each row represents a single observation of COVID-19 data for a particular country on a particular date. The columns contain the following information:

* `date_x`: The date of the observation (in string format).
* `country`: The country for which the observation was made.
* `new_confirmed`: The number of new confirmed cases reported on the date of the observation.
* `new_deceased`: The number of new deaths reported on the date of the observation.
* `new_recovered`: The number of new recoveries reported on the date of the observation.
* `new_tested`: The number of new tests conducted on the date of the observation.
* `cumulative_confirmed`: The cumulative number of confirmed cases up to and including the date of the observation.
* `cumulative_deceased`: The cumulative number of deaths up to and including the date of the observation.
* `cumulative_recovered`: The cumulative number of recoveries up to and including the date of the observation.
* `cumulative_tested`: The cumulative number of tests conducted up to and including the date of the observation.
* `new_persons_vaccinated`: The number of new persons vaccinated on the date of the observation.
* `date`: A datetime object representing the date of the observation.

This dataset contains valuable information about COVID-19 cases, deaths, recoveries, testing, and vaccination across different countries and over time. It can be used to analyze trends in COVID-19 data and to develop models for predicting future trends.

```python
# Set the figure size
plt.figure(figsize=(15, 8))

# Histogram of new confirmed cases
plt.hist(data['new_confirmed'])
plt.xlabel('New Confirmed Cases')
plt.ylabel('Frequency')
plt.title('Distribution of New Confirmed Cases')
plt.show()

# Set the figure size
plt.figure(figsize=(15, 8))

# Line plot of cumulative confirmed cases over time
plt.plot(data['date'], data['cumulative_confirmed'])
plt.xlabel('Date')
plt.ylabel('Cumulative Confirmed Cases')
plt.title('Cumulative Confirmed Cases Over Time')
plt.xticks(rotation=45)


# Format x-axis labels to show only months
# THIS WAS DONE BECUASE OF THE CHALLENGE OF TIGHTLY-PACKED DATES AT THE X-AXIS..
#..of the charts.
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=45)

plt.show()
```

- The global cumulative confirmed cases of COVID-19 have been increasing over time, with a sharp increase in cases in 2020.
- The peak of the pandemic was in June 2021, when there were over 400 million confirmed cases globally.

```python
# Set the figure size to 15 inches wide and 8 inches high
plt.figure(figsize=(15, 8))

# Grouping the data by 'date_x' and calculating the sum of daily new confirmed cases for each date
daily_confirmed = data.groupby('date')['new_confirmed'].sum()

# Plotting a line chart of daily new confirmed cases over time
plt.plot(daily_confirmed.index, daily_confirmed.values)

# Setting the x-axis label as 'Date'
plt.xlabel('Date')

# Setting the y-axis label as 'New Confirmed Cases'
plt.ylabel('New Confirmed Cases')

# Setting the title of the plot as 'Daily New Confirmed Cases Over Time'
plt.title('Daily New Confirmed Cases Over Time')

# Rotating the x-axis labels by 45 degrees for better readability
plt.xticks(rotation=45)

# Setting the major tick locator for the x-axis to show ticks at the beginning of each month
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

# Setting the major tick formatter for the x-axis to display month and year (e.g., Jan 2020)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

# Rotating the x-axis labels again after applying the formatter to prevent overlapping
plt.xticks(rotation=45)

# Displaying the plot
plt.show()
```

* The number of daily new confirmed COVID-19 cases has been increasing steadily over the past few months. This is likely due to the spread of the Omicron variant, which is more transmissible than previous variants.
* The peak of the pandemic was in January 2022, when there were over 200,000 new confirmed cases per day. However, the number of new cases has since declined, and there were only around 100,000 new cases per day in November 2022.
* The chart shows that the number of new cases is still relatively high, even though the pandemic has been ongoing for over two years. This suggests that the virus is still circulating and that it is important to take precautions to prevent the spread of COVID-19.

```python
# Calculate the correlation matrix for the DataFrame 'data'
# The correlation matrix shows the linear relationship between pairs of columns in the DataFrame
corr = data.corr()

# The 'corr' DataFrame will have a shape of (n, n), where n is the number of columns in 'data'
# The values in the correlation matrix will range from -1 to 1:
# -1 indicates a perfect negative linear correlation
# 1 indicates a perfect positive linear correlation
# 0 indicates no linear correlation between the columns

# The correlation matrix will be symmetric, as the correlation between A and B is the same as between B and A

# The diagonal elements of the correlation matrix will be 1, as each column is perfectly correlated with itself

# The 'corr' DataFrame will contain NaN (Not a Number) for non-numeric columns or when there is no variability in the data
# For example, if a column contains all zeros, the correlation with other columns will be NaN

# Note: If the 'data' DataFrame contains non-numeric columns (e.g., object type or categorical data),
# the correlation will only be calculated for numeric columns, and non-numeric columns will be excluded from the 'corr' DataFrame.
```

```python
# Set the figure size
plt.figure(figsize=(15, 8))

# Create a heatmap using seaborn
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt=".2f", linewidths=2)

# Add title and adjust the layout (optional)
plt.title("Correlation Heatmap")
# plt.tight_layout()

# Display the heatmap
plt.show()
```

Analyzing the correlation matrix provided in the dataset to gain insights into the relationships between the variables. Each cell in the matrix represents the correlation coefficient between two variables. A correlation coefficient ranges from -1 to 1, where:
- 1 indicates a perfect positive correlation (as one variable increases, the other increases proportionally),
- -1 indicates a perfect negative correlation (as one variable increases, the other decreases proportionally), and
- 0 indicates no correlation (the variables are independent of each other).

Interpreting the correlation matrix:

1. **"new_confirmed"** and **"new_deceased"** have a positive correlation of approximately 0.54. This suggests that as the number of new confirmed cases increases, the number of new deaths also tends to increase globally.

2. **"new_confirmed"** and **"new_tested"** have a positive correlation of about 0.58. This indicates that higher testing rates are associated with higher numbers of new confirmed cases, which is expected since more tests can identify more positive cases.

3. **"new_confirmed"** and **"new_recovered"** have a positive correlation of around 0.09. While there is a weak positive relationship, it's important to note that this correlation is not very strong.

4. **"new_deceased"** and **"new_tested"** have a positive correlation of approximately 0.54. This implies that higher testing rates are associated with higher numbers of new deaths, possibly because more tests reveal more positive cases, leading to a higher number of deaths.

5. **"cumulative_confirmed"** and **"cumulative_deceased"** have a strong positive correlation of about 0.92. As the total number of confirmed cases increases over time, the total number of deaths also tends to increase, indicating a strong association between the two variables.

6. **"cumulative_confirmed"** and **"cumulative_tested"** have a positive correlation of around 0.56. This suggests that as the number of total confirmed cases increases, more tests are conducted, which helps identify and confirm more cases.

7. **"cumulative_confirmed"** and **"new_tested"** have a positive correlation of approximately 0.58. This reiterates the relationship observed between new confirmed cases and the number of tests conducted.

8. **"new_persons_vaccinated"** and all the other variables have weak negative correlations ranging from approximately -0.14 to -0.19. This indicates that as the number of daily vaccinations increases, there is a slight decrease in the number of new confirmed cases, new deaths, and new tests conducted. However, the correlations are not very strong, so other factors may play a more significant role in these trends.

9. **"cumulative_recovered"** and **"new_recovered"** have a positive correlation of around 0.47. This suggests that as the number of daily recovered cases increases, the total number of recovered cases over time also tends to increase.

We highlighted positive associations between confirmed cases, testing rates, and deaths. The negative correlations between daily vaccinations and various COVID-19 metrics indicate a potential positive impact of vaccinations during the pandemic.

```python
# Set the figure size to 18 inches wide and 10 inches high
plt.figure(figsize=(18, 10))

# Get the unique list of countries from the 'country' column in the DataFrame 'data'
countries = data["country"].unique()

# Uncomment the following line if you need to calculate daily new cases
# Calculate the daily new cases by subtracting the previous day's cumulative confirmed cases from the current day's new confirmed cases
# data["daily_new_cases"] = data["new_confirmed"] - data["cumulative_confirmed"].shift(1)

# Plot the daily new cases for each country using a line plot
for country in countries:
    # Filter the data for each country using boolean indexing
    country_data = data[data["country"] == country]
    # Plot the 'date' column on the x-axis and the 'new_confirmed' column on the y-axis
    plt.plot(country_data["date"], country_data["new_confirmed"], label=country)

# Set the x-axis label as 'Date'
plt.xlabel("Date")

# Set the y-axis label as 'Daily new cases'
plt.ylabel("Daily new cases")

# Set the title of the plot as 'Daily new cases in selected countries'
plt.title("Daily new cases in selected countries")

# Show a legend with country names to identify each line
plt.legend(countries)

# Format the x-axis labels to show only months and years
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
